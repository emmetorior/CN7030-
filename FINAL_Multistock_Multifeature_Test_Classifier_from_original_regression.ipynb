# **SET UP LABELS**

Populate the label column with values as follows:
Either Neutral, or Upward / Downward direction with a small, medium or large move, giving us NEUTRAL, DOWN_SMALL, DOWN_MEDIUM, DOWN_LARGE,  UP_SMALL, UP_MEDIUM, UP_LARGE

UP: +      DOWN: -
NEUTRAL: +- 0.3%
SMALL > 0.3% < 0.6
MEDIUM > 0.6% < 0.9%
LARGE > 0.9%

```python
# Create label column derived from price change percentage and populate it
df_qqq = df_qqq.withColumn(
    "price_change_label", when(abs(col("price_change_percentage")) <= 0.3, "NEUTRAL")\
    .when((abs(col("price_change_percentage")) > 0.3) & (abs(col("price_change_percentage")) <= 0.6), "SMALL")\
    .when((abs(col("price_change_percentage")) > 0.6) & (abs(col("price_change_percentage")) <= 0.9), "MEDIUM")\
    .when(abs(col("price_change_percentage")) > 0.9, "LARGE")\
    .otherwise("NEUTRAL")  # Handle unexpected values
)

# Add "UP" or "DOWN" prefix based on the sign of price_change_percentage
df_qqq = df_qqq.withColumn(
    "price_change_label",
    when(col("price_change_percentage") > 0, concat(lit("UP-"), col("price_change_label")))
    .otherwise(concat(lit("DOWN-"), col("price_change_label")))
)
```

**Decision Tree Classifier**

Decision trees are useful for classification and regression. They often perform weakly on unseen data, because they can sometimes overfit the training, especially with many branches,
plus they can have a bias in favour of the majority and overlook the minority. They are more suitable to data with linear relationships.
Decision making with decision trees is greedy, with a preference for selecting the next best option at each step.
This is often not optimal as there may be shorter paths to the end goal that are hidden behind other branches.
They are sensitive to small data variations and hyperparameters need to be carefully chosen to improve the performance, or the usage of forest methods, like Random Forest can create variations in trees that ultimately perform much better.
The gini impurity allows us to set a probability of impurity (non-homogeneity) in the decision making. The lower the value the more homogenous the data


Again, once the features are combined into the output column, we pass them to the indexer to fit do a 70% - 30% split on the training data versus the test data, and then pass to the classifier to train. Then tested and the predictions are evaluated.

Map the labels to return values. Stock price move will be Neutral, Up or Down and the scale of the move will be small medium or large. Instead of using two different sets of labels, we can combine into one.

```python
# Define the mapping function
def map_prediction_to_label(prediction):
    temp = prediction
    pred = np.double(prediction)
    if (temp  < 0.3 ) and( temp >- 0.3):
      temp = "NEUTRAL"
    elif (temp>0.3):
      temp = "UP-"
    elif (temp<-0.3):
      temp = "DOWN-"

    tmpPred = np.abs(pred)

    if tmpPred < 0.3:
        tmpPredDesc = ''
    elif tmpPred < 0.6:
        tmpPredDesc = 'SMALL'
    elif tmpPred < 0.9:
        tmpPredDesc = 'MEDIUM'
    else:
        tmpPredDesc = 'LARGE'

    return temp + tmpPredDesc

# Register the function as a UDF
map_prediction_udf = udf(map_prediction_to_label, StringType())

# Use the UDF in the DataFrame
results_df = predictions.select(
    'open',
    'close',
    'price_change_percentage',
    'price_change_label',
    'prediction',
    map_prediction_udf(col('prediction')).alias('predicted_label')
)

# Display results
results_df.show()
```

A Brief note about:

**Random Forest - Predicts target / Classifies**
(Supervised) Can handle a high number of dimensions, is slower to overfit (because it's a forest of trees, not just one) and it's accurate and robust but computationally heavy.

**K-Means - groups data into clusters**
(Unsupervised) Easy to implement, computationally light, handles large data easily, but sensitive to cluster numbers, outliers and non linear data.


```python
print(f"Stock data shape after cleaning: {df_qqq.shape}")
# Visualize the features
plt.figure(figsize=(12, 8))
plt.plot(df_qqq.index, df_qqq['Daily_Percent_Change'], label='Daily Percent Change')
#plt.plot(df_qqq.index, df_qqq['Volume'], label='Volume (scaled)', alpha=0.5)
plt.plot(df_qqq.index, df_qqq['MA20'], label='MA20', alpha=0.8)
plt.plot(df_qqq.index, df_qqq['MA50'], label='MA50', alpha=0.8)
plt.fill_between(df_qqq.index, df_qqq['BB_upper'], df_qqq['BB_lower'], color='gray', alpha=0.2, label='Bollinger Bands')
plt.legend()
plt.title(f"{ticker} Feature Visualization")
plt.show()

# Prepare data for clustering
features = df_qqq[['Daily_Percent_Change', 'volume', 'MA20', 'MA50', 'BB_upper', 'BB_lower']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# K-Means Clustering
kmeans = KMeans(n_clusters=10, random_state=42)
df_qqq['Cluster_KMeans'] = kmeans.fit_predict(scaled_features)

#
df_qqq['Target'] = np.where(df_qqq['Daily_Percent_Change'] > 0, 1, 0)

# Classifier
rf = RandomForestClassifier(random_state=42)
rf.fit(scaled_features, df_qqq['Target'])

# predict
df_qqq['Cluster_RF'] = rf.predict(scaled_features)

# Evaluate clustering using silhouette score
silhouette_avg = silhouette_score(scaled_features, df_qqq['Cluster_KMeans'])
print(f"Silhouette Score for K-Means Clustering: {silhouette_avg:.2f}")

from sklearn.cluster import KMeans

# kmeans - elbow calculation
k_range = range(1, 8)
inertia = []
for k in k_range:
  #kmeans = KMeans(n_clusters=k, random_state=99)
  #kmeans.fit(scaled_features)
  kmeans = KMeans(n_clusters=k, random_state=42).fit(scaled_features)
  inertia.append(kmeans.inertia_)
```

Elbow Method

We loop through the number of clusters to be evaluated, so for each k value we create a KMeans model and fit it to the features.
The algorithm works on the WCSS which is the sum of the distances squared between each data point and it's cluster center and these values are stored in a list. (In our example the "inertia" variable)
We can see the WCSS on the elbow plot for each number of clusters k and the knee of the curve shows the optimal

```python
plt.figure(figsize=(8, 6))
plt.plot(k_range, inertia, marker='o', linestyle='-')
plt.plot(k_range, inertia, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia (aka WCSS)')
plt.title('Elbow Method For choosing the optimal k number of clusters')
plt.show()
```

From here, we can visually see that about three clusters might be optimal as that's the knee of the curve
